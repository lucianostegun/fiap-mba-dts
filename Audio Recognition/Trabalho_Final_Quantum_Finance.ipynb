### Função para capturar o áudio do usuário

```python
import speech_recognition as SpeechRec

MICROPHONE_INDEX = 2
recognizer = SpeechRec.Recognizer()

def record_user_input():
  with SpeechRec.Microphone(MICROPHONE_INDEX) as inputDevice:
        print("Inicie a fala para captura do áudio")
        recognizer.adjust_for_ambient_noise(inputDevice,duration=1)
        resp = recognizer.listen(inputDevice)
        return resp
```

### Função para analizar o áudio do usuário e converter para texto

```python
def analyse_input(input):
  try:
    return recognizer.recognize_google(input, language="pt-br")
  except SpeechRec.UnknownValueError:
      return "Google Speech Recognition could not understand audio"
  except SpeechRec.RequestError as e:
      return "Could not request results from Google Speech Recognition service; {0}".format(e)
```

### Função para identificação da opção selecionada

```python
import difflib

def identify_option(input_text, threshold=0.6):
    options = [
      "Consulta ao saldo da conta",
      "Simulação de compra internacional",
      "Falar com um atendente",
      "Sair do atendimento. Lique sempre que precisar"
    ]

    """
    Identifica a opção mais provável com base no texto de entrada.

    Parâmetros:
    - input_text: O texto de entrada a ser comparado.
    - threshold: Limiar de similaridade para considerar uma correspondência.

    Retorna:
    - A opção mais provável ou False se nenhuma opção atender ao limiar.
    """
    best_match = None
    best_score = 0

    for idx, opcao in enumerate(options):
        score = difflib.SequenceMatcher(None, input_text.lower(), opcao.lower()).ratio()

        if score > best_score:
            best_score = score
            best_match = idx + 1

    if best_score >= threshold:
        return best_match
    else:
        return 0

# Exemplo de uso


user_input = "Eu quero consultar o meu saldo"
identified_option = identify_option(user_input)
print(identified_option)

if identified_option:
    print(f"A opção mais provável é: {identified_option}")
else:
    print("Não foi possível identificar uma opção.")
```

### Função principal de início do serviço

```python
from IPython.display import Audio, display
def quantum_finance_phone_service():
  MAX_ATTEMPTS = 3
  attempts = 0;
  display(Audio(filename="saudacao.mp3", autoplay=True))

  while attempts < MAX_ATTEMPTS:
    attempts = attempts + 1
    display(Audio(filename="opcoes.mp3", autoplay=True))

    input = record_user_input()
    speech_to_text = analyse_input(input)
    print(f"Fala do usuário: {speech_to_text}")
    option = identify_option(speech_to_text);
    print(f"Opção identificada: {option}")

    if option > 0:
      audio_file = f"opcao-{option}.mp3"
      print(f"Reproduzindo {audio_file}")
      display(Audio(filename=audio_file, autoplay=True))
      break;
    else:
      display(Audio(filename="nao-identificado.mp3", autoplay=True))

      if attempts >= MAX_ATTEMPTS:
        display(Audio(filename="finalizando-atendimento.mp3", autoplay=True))

quantum_finance_phone_service()
```
