# Exercício

Realizar o Data Inputation utilizando o dataset abaixo para os seguintes casos:

- Valor arbitrário: -100.000.000
- Mediana
- Média
- End of Tail (em +3sigma)
- KNN Inputer

**Dica**: Cuidado para não utilizar o mesmo dataset nos diferentes exercícios

```python
# knn imputation transform for the horse colic dataset
from numpy import isnan
from pandas import read_csv

# Carrega  dataset
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/horse-colic.csv'
df = read_csv(url, header=None, na_values='?')

df
```

```python
# Valor arbitrário: -100.000.000
dataset = df.copy()
dataset.fillna(-100000000)
```

```python
# Mediana
dataset = df.copy()
for col in dataset:
    dataset[col].fillna(dataset[col].median(), inplace=True)

dataset
```

```python
# Média

dataset = df.copy()
for col in dataset:
    dataset[col].fillna(dataset[col].mean(), inplace=True)

dataset
```

```python
# End of Tail (em +3sigma)

dataset = df.copy()
for col in dataset:
    # Calculate the mean and standard deviation
    mean = dataset[col].mean()
    std_dev = dataset[col].std()
    

    # Calculate the threshold for outliers (sigma 3)
    threshold = mean + 3 * std_dev
    dataset[col].fillna(threshold, inplace=True)

dataset
```

```python
# KNN Inputer
import numpy as np
import pandas as pd
from sklearn.impute import KNNImputer

dataset = df.copy()

# Create a KNNImputer with the desired number of neighbors (e.g., 3)
knn_imputer = KNNImputer(n_neighbors=3)

# Use the KNNImputer to fill NA values
df_imputed = knn_imputer.fit_transform(dataset)

# Convert the result back to a DataFrame
df_imputed = pd.DataFrame(df_imputed, columns=dataset.columns)

df_imputed
```
